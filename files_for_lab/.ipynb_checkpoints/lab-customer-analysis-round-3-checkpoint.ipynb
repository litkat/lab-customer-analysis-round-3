```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv("csv_files/marketing_customer_analysis.csv")
df.head(3)
```

### Show DataFrame info

```python
df.info()
```

### Describe DataFrame

```python
df.describe()
```

### Show a plot of the total number of responses

```python
df['Response'].value_counts().plot(kind = 'bar');

#df['Response'].hist()
#plt.show()
```

### Show a plot of the response rate by the sales channel

```python
sns.countplot(x = "Sales Channel", hue = "Response", data = df);
```

### Show a plot of the response rate by the total claim amount

```python
fig, ax = plt.subplots(figsize= (14,5))
#p1 = df.guery('Response == "Yes"')
#p2 = df.guery('Response == "No"')
p1 = df[df['Response'] == 'Yes']
p2 = df[df['Response'] == 'No']

ax = sns.histplot(p1['Total Claim Amount'], bins = 20, label = "Yes", color = 'r')
ax = sns.histplot(p2['Total Claim Amount'], bins = 20, label = "No", color = 'g', alpha = .3)
plt.xlabel('Total Claim Amount', size = '20')
plt.ylabel('Response', size = '20')

#ax = sns.distplot(p1['Total Claim Amount'], bins = 20, label = "Yes")
#ax = sns.distplot(p2['Total Claim Amount'], bins = 20, label = "No")
plt.legend()
plt.show()
```

### Show a plot of the response rate by income

```python
fig, ax = plt.subplots(figsize= (14,5))

p1 = df[df['Response'] == 'Yes']
p2 = df[df['Response'] == 'No']

ax = sns.histplot(p1['Income'], bins = 10, label = "Yes", color = 'r')
ax = sns.histplot(p2['Income'], bins = 10, label = "No", color = 'grey', alpha = .3)
plt.xlabel('Income', size = '20')
plt.ylabel('Response', size = '20')

plt.legend()
plt.show()
```
